```python
from sklearn.utils import resample

# Function to balance the training dataset
def balance_training_data(train_set, target_col='sentiment'):
    # Separate classes
    negative = train_set[train_set[target_col] == 'negative']
    neutral = train_set[train_set[target_col] == 'neutral']
    positive = train_set[train_set[target_col] == 'positive']
    
    # Determine the maximum class size
    max_size = max(len(negative), len(neutral), len(positive))
    
    # Oversample minority classes
    negative_balanced = resample(negative, replace=True, n_samples=max_size, random_state=42)
    neutral_balanced = resample(neutral, replace=True, n_samples=max_size, random_state=42)
    positive_balanced = resample(positive, replace=True, n_samples=max_size, random_state=42)
    
    # Combine balanced classes
    balanced_data = pd.concat([negative_balanced, neutral_balanced, positive_balanced])
    # Shuffle the dataset
    balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)
    return balanced_data

# Balance the `train_drop` dataset
print("Balancing `train_drop`...")
balanced_train_drop = balance_training_data(train_drop)

# Balance the `train_recode` dataset
print("Balancing `train_recode`...")
balanced_train_recode = balance_training_data(train_recode)

# Save balanced training sets for future use
balanced_train_drop.to_csv("balanced_train_drop.csv", index=False)
balanced_train_recode.to_csv("balanced_train_recode.csv", index=False)

# Verify class distribution after balancing
print("\nClass distribution after balancing:")
print(balanced_train_drop['sentiment'].value_counts())
print(balanced_train_recode['sentiment'].value_counts())
```

# **Tokenization and Lemmatization**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Define tokenization and lemmatization function
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Tokenize and convert to lowercase
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum()]  # Lemmatize and remove non-alphanumeric
    return " ".join(lemmatized_tokens)

# Apply preprocessing to all splits and save results
def preprocess_dataset(dataset, text_column='reviews.text'):
    dataset[text_column] = dataset[text_column].apply(preprocess_text)
    return dataset

print("Tokenizing and lemmatizing `train_drop`...")
train_drop = preprocess_dataset(train_drop)
val_drop = preprocess_dataset(val_drop)
test_drop = preprocess_dataset(test_drop)

print("Tokenizing and lemmatizing `train_recode`...")
train_recode = preprocess_dataset(train_recode)
val_recode = preprocess_dataset(val_recode)
test_recode = preprocess_dataset(test_recode)

# Save preprocessed datasets for future use
train_drop.to_csv("preprocessed_train_drop.csv", index=False)
val_drop.to_csv("preprocessed_val_drop.csv", index=False)
test_drop.to_csv("preprocessed_test_drop.csv", index=False)

train_recode.to_csv("preprocessed_train_recode.csv", index=False)
val_recode.to_csv("preprocessed_val_recode.csv", index=False)
test_recode.to_csv("preprocessed_test_recode.csv", index=False)

print("\nPreprocessed datasets saved successfully!")
```

# **Vectorization applying TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Define the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2), stop_words='english')

# Function to vectorize datasets
def vectorize_datasets(train, val, test, text_column='reviews.text'):
    # Fit TF-IDF on the training set
    X_train = tfidf_vectorizer.fit_transform(train[text_column])
    X_val = tfidf_vectorizer.transform(val[text_column])
    X_test = tfidf_vectorizer.transform(test[text_column])
    
    return X_train, X_val, X_test

print("Vectorizing `dataset_drop`...")
X_train_drop, X_val_drop, X_test_drop = vectorize_datasets(train_drop, val_drop, test_drop)

print("Vectorizing `dataset_recode`...")
X_train_recode, X_val_recode, X_test_recode = vectorize_datasets(train_recode, val_recode, test_recode)
```

# **Models Training and Evaluation**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Function to train, evaluate, and calculate confusion matrix
def train_evaluate_with_confusion_matrix(model, X_train, y_train, X_val, y_val, X_test, y_test, model_name):
    print(f"\nTraining {model_name}...")
    model.fit(X_train, y_train)

    # Evaluate on validation set
    y_val_pred = model.predict(X_val)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    print(f"Validation Accuracy ({model_name}): {val_accuracy:.4f}")

    # Evaluate on test set
    y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    print(f"Test Accuracy ({model_name}): {test_accuracy:.4f}")

    # Classification Report
    print(f"\nClassification Report ({model_name} - Test Set):")
    print(classification_report(y_test, y_test_pred))

    # Confusion Matrix
    print(f"\nConfusion Matrix ({model_name} - Test Set):")
    cm = confusion_matrix(y_test, y_test_pred, labels=['positive', 'neutral', 'negative'])
    print(cm)

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['positive', 'neutral', 'negative'], yticklabels=['positive', 'neutral', 'negative'])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    return model

# Load vectorized features and labels
# Assuming `train_drop` and `train_recode` sentiment labels are in `y_train_drop` and `y_train_recode`
y_train_drop = train_drop['sentiment']
y_val_drop = val_drop['sentiment']
y_test_drop = test_drop['sentiment']

y_train_recode = train_recode['sentiment']
y_val_recode = val_recode['sentiment']
y_test_recode = test_recode['sentiment']

# Train and evaluate models on `dataset_drop`
print("\n--- Training on `dataset_drop` ---")
nb_drop = train_evaluate_with_confusion_matrix(MultinomialNB(), X_train_drop, y_train_drop, X_val_drop, y_val_drop, X_test_drop, y_test_drop, "Naive Bayes")
lr_drop = train_evaluate_with_confusion_matrix(LogisticRegression(max_iter=1000), X_train_drop, y_train_drop, X_val_drop, y_val_drop, X_test_drop, y_test_drop, "Logistic Regression")
svm_drop = train_evaluate_with_confusion_matrix(SVC(), X_train_drop, y_train_drop, X_val_drop, y_val_drop, X_test_drop, y_test_drop, "SVM")
rf_drop = train_evaluate_with_confusion_matrix(RandomForestClassifier(), X_train_drop, y_train_drop, X_val_drop, y_val_drop, X_test_drop, y_test_drop, "Random Forest")

# Train and evaluate models on `dataset_recode`
print("\n--- Training on `dataset_recode` ---")
nb_recode = train_evaluate_with_confusion_matrix(MultinomialNB(), X_train_recode, y_train_recode, X_val_recode, y_val_recode, X_test_recode, y_test_recode, "Naive Bayes")
lr_recode = train_evaluate_with_confusion_matrix(LogisticRegression(max_iter=1000), X_train_recode, y_train_recode, X_val_recode, y_val_recode, X_test_recode, y_test_recode, "Logistic Regression")
svm_recode = train_evaluate_with_confusion_matrix(SVC(), X_train_recode, y_train_recode, X_val_recode, y_val_recode, X_test_recode, y_test_recode, "SVM")
rf_recode = train_evaluate_with_confusion_matrix(RandomForestClassifier(), X_train_recode, y_train_recode, X_val_recode, y_val_recode, X_test_recode, y_test_recode, "Random Forest")
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import LabelEncoder

# Function to train and evaluate XGBoost model
def train_evaluate_xgboost(X_train, y_train, X_val, y_val, X_test, y_test, dataset_name):
    print(f"\n--- Training XGBoost on {dataset_name} ---")

    # Encode string labels into integers
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_val_encoded = label_encoder.transform(y_val)
    y_test_encoded = label_encoder.transform(y_test)

    # Initialize XGBoost model with basic parameters
    xgb_model = XGBClassifier(
        use_label_encoder=False,  # Suppress warning
        eval_metric='logloss',   # Metric for evaluation
        scale_pos_weight=1,      # Placeholder for class balancing
        random_state=42
    )

    # Train the model
    xgb_model.fit(X_train, y_train_encoded)

    # Validation predictions and metrics
    val_predictions = xgb_model.predict(X_val)
    val_f1_score = f1_score(y_val_encoded, val_predictions, average='macro')
    print(f"Validation F1 Macro Score (XGBoost - {dataset_name}):", val_f1_score)

    # Test predictions and metrics
    test_predictions = xgb_model.predict(X_test)
    test_f1_score = f1_score(y_test_encoded, test_predictions, average='macro')
    print(f"Test F1 Macro Score (XGBoost - {dataset_name}):", test_f1_score)

    # Classification report
    print("\nClassification Report (XGBoost - Test Set):\n")
    print(classification_report(y_test_encoded, test_predictions, target_names=label_encoder.classes_))

    # Confusion matrix
    print("Confusion Matrix (XGBoost - Test Set):\n")
    print(confusion_matrix(y_test_encoded, test_predictions))

    return xgb_model

# Train and evaluate XGBoost on dataset_drop
xgb_drop = train_evaluate_xgboost(
    X_train=X_train_drop,
    y_train=y_train_drop,
    X_val=X_val_drop,
    y_val=y_val_drop,
    X_test=X_test_drop,
    y_test=y_test_drop,
    dataset_name="dataset_drop"
)

# Train and evaluate XGBoost on dataset_recode
xgb_recode = train_evaluate_xgboost(
    X_train=X_train_recode,
    y_train=y_train_recode,
    X_val=X_val_recode,
    y_val=y_val_recode,
    X_test=X_test_recode,
    y_test=y_test_recode,
    dataset_name="dataset_recode"
)
```

# **XGBoostClassifier Hyperparameter Tunning**

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.model_selection import GridSearchCV

# Function to encode target labels
def encode_labels(y_train, y_val, y_test):
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_val_encoded = label_encoder.transform(y_val)
    y_test_encoded = label_encoder.transform(y_test)
    print("Label Encoding Mapping:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
    return y_train_encoded, y_val_encoded, y_test_encoded

# Encode labels for dataset_drop
y_train_drop, y_val_drop, y_test_drop = encode_labels(y_train_drop, y_val_drop, y_test_drop)

# Encode labels for dataset_recode
y_train_recode, y_val_recode, y_test_recode = encode_labels(y_train_recode, y_val_recode, y_test_recode)

# Function to train and evaluate XGBoost model with optional GridSearchCV
def train_evaluate_xgboost(X_train, y_train, X_val, y_val, X_test, y_test, dataset_name, use_grid_search=False):
    print(f"\n--- Training XGBoost on {dataset_name} ---")

    if use_grid_search:
        # Define parameter grid for GridSearchCV
        param_grid = {
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.1, 0.2],
            'n_estimators': [50, 100, 200],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0]
        }

        xgb_model = XGBClassifier(
            eval_metric='logloss',
            random_state=42
        )

        # Perform GridSearchCV
        grid_search = GridSearchCV(
            estimator=xgb_model,
            param_grid=param_grid,
            scoring='f1_macro',
            cv=5,
            verbose=1
        )
        grid_search.fit(X_train, y_train)

        # Best parameters and model
        best_params = grid_search.best_params_
        print(f"Best Parameters for XGBoost ({dataset_name}):", best_params)
        xgb_model = grid_search.best_estimator_

    else:
        # Initialize XGBoost model with basic parameters
        xgb_model = XGBClassifier(
            eval_metric='logloss',
            random_state=42
        )

        # Train the model
        xgb_model.fit(X_train, y_train)

    # Validation predictions and metrics
    val_predictions = xgb_model.predict(X_val)
    val_f1_score = f1_score(y_val, val_predictions, average='macro')
    print(f"Validation F1 Macro Score (XGBoost - {dataset_name}):", val_f1_score)

    # Test predictions and metrics
    test_predictions = xgb_model.predict(X_test)
    test_f1_score = f1_score(y_test, test_predictions, average='macro')
    print(f"Test F1 Macro Score (XGBoost - {dataset_name}):", test_f1_score)

    # Classification report
    print("\nClassification Report (XGBoost - Test Set):\n")
    print(classification_report(y_test, test_predictions))

    # Confusion matrix
    print("Confusion Matrix (XGBoost - Test Set):\n")
    print(confusion_matrix(y_test, test_predictions))

    return xgb_model

# Train and evaluate XGBoost on dataset_drop with GridSearchCV
xgb_drop = train_evaluate_xgboost(
    X_train=X_train_drop,
    y_train=y_train_drop,
    X_val=X_val_drop,
    y_val=y_val_drop,
    X_test=X_test_drop,
    y_test=y_test_drop,
    dataset_name="dataset_drop",
    use_grid_search=True
)

# Train and evaluate XGBoost on dataset_recode with GridSearchCV
xgb_recode = train_evaluate_xgboost(
    X_train=X_train_recode,
    y_train=y_train_recode,
    X_val=X_val_recode,
    y_val=y_val_recode,
    X_test=X_test_recode,
    y_test=y_test_recode,
    dataset_name="dataset_recode",
    use_grid_search=True
)
```
